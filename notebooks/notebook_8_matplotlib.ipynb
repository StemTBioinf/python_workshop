```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
```

```python
fdata = pd.read_table('../data/fluidigm_data.tsv', index_col=0)
fdata.shape
```

```python
sns.scatterplot(x='CD34', y='EPCR', data=fdata)
```

```python
plt.scatter(fdata['CD34'].values, fdata['EPCR'].values)
```

```python
fig = plt.figure(figsize=(4,4))
plt.show()
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1, 1, 1)
plt.show()
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1, 1, 1)
ax.plot([0, 1], [0, 1])
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(fdata['CD34'].values, fdata['EPCR'].values)
plt.show()
```

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(1, 2, 1)
ax.scatter(fdata['CD34'].values, fdata['EPCR'].values)
plt.show()
```

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(1, 2, 1)
ax.scatter(fdata['CD34'].values, fdata['EPCR'].values)
ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(fdata['CD48'].values, fdata['CD150'].values)
plt.show()
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(fdata['CD34'].values, fdata['EPCR'].values,
           s=5, c='red', label='comp1')
ax.set_xlabel('CD34', fontsize=16)
ax.set_ylabel('EPCR', fontsize=16)
ax.legend()
ax.set_title('EPCR vs CD34')
plt.show()
```

```python
sns.clustermap(fdata)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
zdata = StandardScaler().fit_transform(fdata)
zdata = pd.DataFrame(zdata, index=fdata.index, columns=fdata.columns)
```

```python
zdata.sum()
```

```python
sns.clustermap(zdata)
```

```python
clustering = sns.clustermap(zdata, figsize=(5,5))
```

```python
clustering
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
Xpca = pca.fit_transform(zdata).T
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(Xpca[0], Xpca[1],
           s=5, c='red')
ax.set_xlabel('PC1', fontsize=16)
ax.set_ylabel('PC2', fontsize=16)
ax.set_title('PCA')
plt.show()
```
